```python
import cv2
import numpy as np
```

```python
capture = cv2.VideoCapture(0)
ret, frame = capture.read () 
while (True):
 
    ret, frame = capture. read ()
    #Converting the captured video frames to apply effects
    grayFrame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    inversion_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    blur_image = cv2.blur(frame, (5,5))
    cartoon_image = cv2.bilateralFilter(frame,5,255,255)
    sketch_image = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    sketch_image = cv2.medianBlur(sketch_image,3)
    gray_image = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    
    
    
    #Applying effects to the respective converted frames
    edges = cv2.adaptiveThreshold(sketch_image,255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,5,5)
    cartoon_image = cv2.bitwise_and (cartoon_image, cartoon_image, mask=edges)
    X_ray_image = cv2.bitwise_not(frame)
    Canny_edge_image = cv2.Canny(frame,100,200)
    thresh, blackAndWhiteFrame = cv2.threshold(grayFrame, 127, 255, cv2.THRESH_BINARY)
    thresh, INVERSION_FRAME = cv2.threshold(inversion_frame,127,255, cv2.THRESH_BINARY)
    
    #resizing the image according to our need  
    # resize () function takes 2 parameters,   
    # the image and the dimensions  
    input_image = cv2.resize(frame, (300, 300))
    # Extracting the height and width of an image  
    rows, cols = input_image. shape [:2] 
    # generating vignette mask using Gaussian  
    # resultant_kernels 
    X_resultant_kernel = cv2.getGaussianKernel(cols,200) 
    Y_resultant_kernel = cv2.getGaussianKernel(rows,200) 
    #generating resultant_kernel matrix  
    resultant_kernel = Y_resultant_kernel * X_resultant_kernel. T 
    #creating mask and normalising by using np. linalg 
    # function 
    mask = 255 * resultant_kernel / np. linalg. norm(resultant_kernel) 
    output = np. copy(input_image) 
    # applying the mask to each channel in the input image 
    for i in range (3): 
        output[:, :, i] = output[:,:,i] * mask
        
        
 
    #resizing the frames by (300,300) to the applied effects by using the method resize ()
    
    resized_bw = cv2.resize(blackAndWhiteFrame, (300,300))
    resized_inv = cv2.resize(INVERSION_FRAME, (300,300))
    resized_cartoon = cv2.resize(cartoon_image, (300,300))
    resized_blur_image = cv2.resize(blur_image, (300,300))
    resized_edges_img = cv2.resize(edges, (300,300))
    resized_xray_image = cv2.resize(X_ray_image, (300,300))
    resized_gray_image = cv2.resize(gray_image, (300,300))
    resized_canny_image = cv2.resize(Canny_edge_image, (300,300))
    
    #Showing all the effects applied to the taken frame by using the method imshow ()
    cv2.imshow('video bw', resized_bw)
    cv2.imshow('video inv', resized_inv)
    cv2.imshow('video original', frame)
    cv2.imshow('video blur', resized_blur_image)
    cv2.imshow('sketch art', resized_edges_img)
    cv2.imshow('cartoon', resized_cartoon)
    cv2.imshow('X-Ray', resized_xray_image)
    cv2.imshow('Gray Image',resized_gray_image)
    cv2.imshow('VIGNETTE', output)
    cv2.imshow('Canny Edge',resized_canny_image)
   
    
 
    if cv2.waitKey(0):
        break
```
